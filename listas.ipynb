# 11 numbers
Para poder organizar una lista de números o de palabras, python lo va a poder organizar con el comando *"sorted"* lo cual lo hara de menor a mayor en el caso de los números y en las palabras lo hara en orden alfabetico, declaramos la nueva variable y en el *"sorted"* ponemos el nombre de la lista original.

```python
numbers = [5, 2, 9, 1, 7]
numbers_organized = sorted(numbers)
print(numbers_organized)
```

# 12 numbers

# 13 numbers
Nos pide hacer un bucle en el cual todos los números de la lista que tengamos sean multiplicado por dos, primero declaramos la nueva variable que va a ser igual a los números multiplicados por 2 y va el *"for"* que especificamos que es para armar secuencias  en listas,rangos o cadenas.Encontes le decimos que va a repetir esa secuencia de multiplicar los números por dos número por número en la lista ese *"in"* le hace la aclaración

```python
numbers = [3, 6, 9, 12, 15]
multiply = [number * 2 for number in numbers]
print(multiply)
```

# 13 words
Lo que nos pide el ejercicio es recorrer una lista de palabras y guardarlas en una nueva lista usando una comprensión de lista. Primero declaramos la lista original con las palabras que queremos, luego creamos una nueva variable con el mismo nombre y usamos un bucle *"for"* dentro de corchetes para ir tomando cada palabra una por una de la lista original, esto es lo que se llama comprensión de lista, y al final imprimimos el resultado.

```python
words = ["python", "list", "exercise", "practice"]
words =[ words for words in words]
print(words)
```

# 15 Average
En este ejercicio nos piden calcular el promedio de una lista de calificaciones. Para esto primero tenemos que importar el módulo *"statistics"* que es una librería de python que nos ayuda a hacer cálculos estadísticos. Luego declaramos la lista con las notas, después creamos una nueva variable *"average"* y usamos el comando *"statistics.mean"* pasándole el nombre de la lista, lo que hace es sumar todos los elementos y dividirlos entre la cantidad total, y por último imprimimos el resultado.

```python
import statistics
grades = [80, 75, 90, 85, 95]
average = statistics.mean(grades)
print(average)
```
