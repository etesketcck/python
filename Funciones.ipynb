#7 función *"is_even"*
En esta función denominada *"is_even"* queremos saber si el número que coloquemos sea par y nos diga si es falso o verdadero, entonces partimos por el comando de las funciones y el nombre del inicio y declaramos en el parentesis que seran solo números, ahora ponermos el *"return"* y que ponga el número y que es número al momento de divirlo entre 2 gracias al signo de *"%"* este no deje ningún residuo, indicando que es par, ya que si deja residuo no seria par, y verificamos poniendo números pares e impares en el *"print"* verificando si esa haciendo bien en verificar la veracidad de los números pares.  

```python
def is_even(nunber):
  return nunber % 2 == 0
print(is_even(5))
print(is_even(6))
print(is_even(8))
```

# 14  función *"average"*
Declaramos la función y en el parentesis ponemos que vamos a trabajar con los números, le ponemos *"return"* y le decimos que sume los números con el comando *"sum"* entre parentesis *"numbres"* y que esa suma la divida entre el dato de cuantos números tiene que se pone con el *"len(number)"* y por ultimo imprimos el conjunto de números que quereamos que saco la media o tambien se puede importar los comando de estadistica de la siguiente manera *"import statistics"* para que podamos usar lso comando de ese paquetre y en vez de poner todo eso de *"sum y len"* le pongamos *"statistics.mean"* y saque la media más facil.


```python
def average(numbers):
 return sum(numbers)/len(numbers)
print(average([1,2,3,4,5]))
print(average([6,7,8,934,24]))
```

```python
import statistics
def calculate_statistics(numbers):
  return statistics.mean(numbers)
print(calculate_statistics([1,2,3,4,5]))
print(calculate_statistics([6,7,8,934,24]))
```

# 15  *"función factorial"*
En este caso se hace los mismo declaramos la función y le ponemos entre parentesi *"n"* y le decimos que si *"n"* es 0 o 1 devuelva 1 y se detiene esto es necesario para que no se vuelva infinito el código y que no de error o se demore en procesar todo esos números, ahora que *"n"* es mayor que 1, la función se llama asi con *"n-1"* hasta llegar al caso base, haci quedaria las factoriales , y por ultimo ponemos el print para que lo haga las facotriales dando el resultado

```python
def factorial(n):
 if n == 0 or n == 1:
     return 1
 return n * factorial(n - 1)
print(factorial(5))
print(factorial(7))
```
