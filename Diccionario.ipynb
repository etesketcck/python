# 7 fruits
Imprimir solo las llaves del diccionario, se hace primero poniendo el diccionario original para despies poner el comando *"for"* despues declaramos el nuevo nombre de la variable el cual le pusimos  *"names_of_fruits"*, despues colocamos otro comando que es *"in"* y pondremos el nombre del diccionario y aqui cambiar algo, ponemo punto y tenemos dos opciones *"keys"* o *"values"*. En este caso sera el *"keys"*, y le decimos a python que imprima la  nueva variable que acabamos de declarar.

```python
fruits = {"apple": 3, "banana": 5, "orange": 2}
for names_of_fruits in fruits.keys():
    print(names_of_fruits)
```

# 8 fruits
Imprimir solo los los valores, es lo mismo que en el caso anterior sino que tenemos que declarar la cantidad con otros nombre y poner en vez de *"keys"* tenemos que colocar el comando *"values"* despues del *"for"*, va el nombre de la variable que vamos a declara que es  *"amount_of_fruit"* que sigue el *"in"* y por ultimo el *"fruits.values"*.

```python
fruits = {"apple": 3, "banana": 5, "orange": 2}
for amount_of_fruit in  fruits.values():
  print(amount_of_fruit)
```

# 9 fruits
Ahora para poner todo los comandos y elementos que acompaña el diccionario, primero ponemos el diccionario original para despues poner de nuevo el *"of"* para luego declarar las dos variables que queremos ver reflejadas que sera *"names_of_fruits"* y *"amount_of_fruit"*, para poner  *"in"* y por ultimo poner el nombre de la variable con un punto el comando "*items*" que representara todos las llaves y los elementos.

```python
fruits = {"apple": 3, "banana": 5, "orange": 2}
for names_of_fruits,amount_of_fruit in fruits.items():
  print(names_of_fruits,amount_of_fruit)
```

#12 numbers
Realizar la suma total de las variables, lo  que se hace es poner el diccionario original, se declara como nuevo *"suma"*  para luego imprimirla, ponemos el signo igual, luego el comando *"sum"*, dentro del parentesis poner el nombre del diccionario y el *".values"* para  que solo tome los valores de la llave y haga una suma total.

# 14 countries
Ahora para poner todo las llaves  y elementos que acompaña el diccionario, primero ponemos el diccionario original para despues poner de nuevo el *"of"* para luego declarar las dos variables que queremos ver reflejadas que sera *"country"* y *"capital"*, para poner  *"in"* y por ultimo poner el nombre de la variable con un punto el comando "*items*" que representara todos las llaves y los elementos. Para despues imprimir el número.

```python
countries = {"Colombia": "Bogota", "France": "Paris", "Japan": "Tokyo"}
for country, capital in countries.items():
    print(country,capital)
```
